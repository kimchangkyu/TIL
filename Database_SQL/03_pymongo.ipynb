## pymongo

- python에서 MongoDB 작업을 위한 도구
- 설치 : pip install pymongo==2.8.1

```python
import pymongo
import pandas as pd
```

- 직방 데이터 사용
- zigbang.py 업로드 후

```python
import zigbang as zb
```

#### 1. 데이터 베이스 서버 연결

```python
client = pymongo.MongoClient('mongodb://13.124.41.91:27017/') # public ip:port number
client
```

#### 2. 데이터 베이스와 컬렉션 생성

```python
zigbang = client.crawling.zigbang # server.database.collection
zigbang
```

```python
# 크롤링
items = zb.main("성수동")
len(items)
```

```python
items[0] # 딕셔너리 데이터
```

#### insert

```python
ids = zigbang.insert(items)
len(ids)
```

#### 3. 데이터 프레임으로 저장

```python
sungsoo_df = pd.DataFrame(items)
columns = ["_floor", "address1", "rent", "size", "deposit", "options"] # 필요한 정보만
sungsoo_df = sungsoo_df[columns]
sungsoo_df.tail(2)
```

```python
# 객체 생성
zigbang2 = client.crawling.zigbang2 
```

```python
items = sungsoo_df.to_dict("records")
```

```python
items[0]
```

```python
ids = zigbang2.insert(items)
```

#### 4. 데이터 find()

```python
results = zigbang2.find()
list(results)
```

```python
QUERY = {"rent": {"$lte": 50}}
results = zigbang2.find(QUERY)
pd.DataFrame(results).tail()
```

#### 5. 컬렉션 삭제

```python
client.crawling.drop_collection("zigbang")
```

#### 6. 데이터 베이스 삭제

```python
client.drop_database("crawling")
```
